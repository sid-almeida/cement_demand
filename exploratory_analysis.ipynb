**1. Importing the libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn
import seaborn as sns
import datetime as dt
```

**2. Importing the dataset**

```python
data = pd.read_csv('explore.csv')
data = data.drop(['Unnamed: 0'], axis=1)
data.head(10)
```

**3. Variable Distribution using seaborn colorful charts**

```python
sns.set_style('whitegrid')
data.hist(figsize=(20,20))
```

**4. Linechart of each variable using seaborn**

```python
sns.set_style('whitegrid', {'axes.grid' : False})
data.plot(kind='line', subplots=True, layout=(4,4), sharex=False, sharey=False, figsize=(20,20))
```

**5. Correlation Matrix**

```python
corr = data.corr()
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm', linewidths=2, linecolor='black')
```

**6. Scatterplot Matrix**

```python
sns.pairplot(data, hue='demand', palette='coolwarm', diag_kind='kde')
```

**7. Boxplots em subplot com sns e for loop**

```python
fig, axes = plt.subplots(3, 3, figsize=(20,20))
sns.boxplot(ax=axes[0, 0], data=data['demand'], orient='h', palette='coolwarm')
sns.boxplot(ax=axes[0, 1], data=data['production'], orient='h', palette='coolwarm')
sns.boxplot(ax=axes[0, 2], data=data['sales '], orient='h', palette='coolwarm')
sns.boxplot(ax=axes[1, 0], data=data['population'], orient='h', palette='coolwarm')
sns.boxplot(ax=axes[1, 1], data=data['gdp'], orient='h', palette='coolwarm')
sns.boxplot(ax=axes[1, 2], data=data['disbursement'], orient='h', palette='coolwarm')
sns.boxplot(ax=axes[2, 0], data=data['interest_rate'], orient='h', palette='coolwarm')
fig.delaxes(axes[2, 1])
fig.delaxes(axes[2, 2])
```
